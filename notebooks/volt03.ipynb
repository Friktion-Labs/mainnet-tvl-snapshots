```python
query = """
query MyQuery {
  solana(network: solana) {
    transfers(
      externalProgramId: {is: "VoLT1mJz1sbnxwq5Fv2SXjdVDgPXrb9tJyC8WpMDkSp"}
      receiverAddress: {is: "C2k7bZg3iBaqGM4xXzJiX4VADa2GJ6ZUhdQV76A5dB3Z"}
      time: {after: "2022-04-21T21:00:00+00:00"}
    ) {
      currency {
        address
        name
        symbol
        tokenType
        tokenId
      }
      sender {
        address
        mintAccount
        type
      }
      receiver {
        address
        mintAccount
        type
      }
      transferType
      amount
      transaction {
          signature
          signer
      }
      block {
        timestamp {
          iso8601
        }
      }
    }
  }
}

"""
```

```python
x = Query("BQYCaXaMZlqZrPCSQVsiJrKtxKRVcSe4")
df = x.run(query, to_df=True)
```

```python
df.sort_values("BlockTimestampIso8601")
```

```python
df.sort_values("BlockTimestampIso8601").TransactionSignature.values
```

```python
df.sort_values("BlockTimestampIso8601")[["TransactionSigner", "TransactionSignature", "BlockTimestampIso8601"]][17:]
```

```python
df["isMint"] = df.groupby("TransactionSignature").Transfertype.transform(lambda x: x=="mint")
```

```python
df["mintAmount"] = df.apply(lambda x: x.Amount if x.isMint else 0, axis=1)
```

```python
df["nonAmount"] = df.apply(lambda x: x.Amount if not x.isMint else 0, axis=1)
```

```python
df.groupby("TransactionSignature").mintAmount.max()/df.groupby("TransactionSignature").nonAmount.max()
```

```python
df.groupby("TransactionSignature").mintAmount.max()/df.groupby("TransactionSignature").nonAmount.max().sort_values()
```
